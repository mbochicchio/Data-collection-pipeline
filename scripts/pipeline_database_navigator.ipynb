```python
import duckdb
conn = duckdb.connect("../data/pipeline.duckdb")
```

```python
conn.execute("SHOW TABLES").fetchall()
```

```python
projects_list = conn.execute("SELECT * FROM projects").fetchall()
count = 0
for project in projects_list:
    print(project)
    count += 1
    if count == 10:
        break
```

```python
results = conn.execute("""
    SELECT language, COUNT(*) AS count
    FROM projects
    GROUP BY language
    ORDER BY count DESC
""").fetchall()

print(f"{'Language':<20} {'Count':>6}")
print("-" * 28)
for language, count in results:
    print(f"{language:<20} {count:>6}")

print(f"\nTotal projects: {sum(r[1] for r in results)}")
```
